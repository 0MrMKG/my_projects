```python
def mkdir(path):
    judge = os.path.exists(path)
    if judge:
        print("已经存在{}".format(path))
        pass
    else:
        print("创建了{}".format(path))
        os.makedirs(path)

def nii2img(filepath,imgfile):

    filenames = os.listdir(filepath)

    for f in filenames:
        img_path = os.path.join(filepath,f)
        img = nib.load(img_path)
        img_fdata = img.get_fdata()
        img_fdata = img_fdata[:,:,:,0]

        fname = f.replace('.nii','')
        img_f_path = os.path.join(imgfile,fname)

        mkdir(img_f_path)

        (x, y, z) = img_fdata.shape
        for i in range(int(z/8),int(7*z/8)):
            slice = img_fdata[:,:,i]
            slice_min = np.min(slice)
            slice_max = np.max(slice)
            slice_normalized  = (slice-slice_min)/(slice_max-slice_min)
            slice_scaled = (slice_normalized * 255).astype(np.uint8)

            slice_image = Image.fromarray(slice_scaled)
            slice_image.save(os.path.join(img_f_path,'{}.jpg'.format(i)))

```
